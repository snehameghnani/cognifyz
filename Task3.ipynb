### step 1: importing the libraries

### step 2: loading the dataset

```python
df= pd.read_csv(r'C:\machine-learning\cognifyz-internship\Dataset .csv')
```

### step 3: handling missing values

```python
print("\nMissing values per column:")
df.isnull().sum()
```

```python
df.dropna(inplace=True)
print("\nDataset shape after dropping missing values:", df.shape)
```

### step 4:feature engineering-extracting features

```python
df['Name Length'] = df['Restaurant Name'].astype(str).apply(len)
```

```python
df['Address Length'] = df['Address'].astype(str).apply(len)
```

```python
df[['Restaurant Name','Name Length', 'Address', 'Address Length']].head()
```

### step 5: encoding categorical variables

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("\nColumns after cleaning:")
print(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_columns = ['restaurant_id', 'restaurant_name', 'country_code', 'city', 'address',
       'locality', 'locality_verbose', 'longitude', 'latitude', 'cuisines',
       'average_cost_for_two', 'currency', 'has_table_booking',
       'has_online_delivery', 'is_delivering_now', 'switch_to_order_menu',
       'price_range', 'aggregate_rating', 'rating_color', 'rating_text',
       'votes']
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column].astype(str))
```

```python
print("\nDataset after encoding categorical variables:")
df.head(10)
```

### step 6: creating new features for table booking and online delivery

```python
df['Table_Bookings'] = df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df['Online_delivery'] = df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)
```
