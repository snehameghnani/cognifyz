# Data Exploration and Preprocessing

### step 1: importing libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### step2: loading the dataset

```python
df= pd.read_csv(r'C:\machine-learning\cognifyz-internship\Dataset .csv')
```

```python
df.head()
```

### step 3: handling missing values

```python
print("Dataset shape:", df.shape)
```

```python
print("\nMissing values per column:")
df.isnull().sum()
```

```python
df.dropna().shape
```

```python
df.dropna(inplace=True)
print("\nDataset shape after dropping missing values:", df.shape)
```

```python
df.dropna(axis=1)
```

### step 4: conversion of datatype

```python
binary_cols = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
for col in binary_cols:
    df[col] = df[col].map({'Yes': True, 'No': False})
    
print("\nUpdated data types:")
print(df.dtypes)
```

### step 5: aggeregate rating distribution

```python
rating_dis = df['Aggregate rating'].value_counts().sort_index()
print("\nAggregate rating distribution:")
print(rating_dis)
```

### step 6: plotting the distribution

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=rating_dis.index, y=rating_dis.values, palette='coolwarm', hue=rating_dis.index, legend=False)
plt.title("Original Aggregate Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### step 7: class imbalance data

```python
total = len(df)
zero_ratings = (df['Aggregate rating'] == 0.0).sum()
non_zero = total - zero_ratings
print(f"\nTotal entries: {total}")
print(f"Zero ratings: {zero_ratings}")
print(f"Non-zero ratings: {non_zero}")
print(f"Zero rating percentage: {zero_ratings / total * 100:.2f}%")
```

### step 8: downsampling zero-ratings

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.utils import resample
```

```python
df_zero = df[df['Aggregate rating'] == 0.0]
df_nonzero = df[df['Aggregate rating'] != 0.0]
mean_classsize = int(df_nonzero['Aggregate rating'].value_counts().mean())

df_zerodown = resample(df_zero, replace=False, n_samples=mean_classsize, random_state=42)
df_balanced = pd.concat([df_nonzero, df_zerodown])
```

### step 9: plotting the balanced distribution

```python
balanced_dis = df_balanced['Aggregate rating'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=balanced_dis.index, y=balanced_dis.values, palette='coolwarm', hue=balanced_dis.index, legend=False)
plt.title("Balanced Aggregate Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
