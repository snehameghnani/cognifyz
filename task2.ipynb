# Descriptive Analysis


### step 1: importing the libraries


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

### step2: loading the dataset

```python
#load the dataset
df= pd.read_csv(r'C:\machine-learning\cognifyz-internship\Dataset .csv')
print("shape of dataset:", df.shape)
```

```python
df.describe()
```

### step 3: calculating basic statistical measures

```python
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Variance:\n", df.var(numeric_only=True))
print("Minimum:\n", df.min(numeric_only=True))
print("Maximum:\n", df.max(numeric_only=True))
```

### step4:exploring the distribution of categorical variables
#### (country code, city, cuisine)

```python
#For Top 10 Countries
plt.figure(figsize=(10, 6))
df['Country Code'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Country Codes by Restaurant Count')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.show()

#For Top 10 Cities
plt.figure(figsize=(12, 6))
df['City'].value_counts().head(10).plot(kind='bar', color='yellow')
plt.title('Top 10 Cities by Restaurant Count')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

```

```python

#For Top 10 Cuisines
from collections import Counter
cusine_series=df['Cuisines'].dropna().apply(lambda x: x.split(',')[0])
top_cuisines = cusine_series.value_counts().head(10)

print("Top 10 Cuisines:\n", top_cuisines)

plt.figure(figsize=(10, 6))
top_cuisines.plot(kind='bar', color='green')
plt.title('Top 10 Cuisines by Restaurant Count')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
#For Top 10 cities with most number of Restaurants

top_cities = df['City'].value_counts().head(10)
print("Top 10 Cities with Most Restaurants:\n", top_cities)

plt.figure(figsize=(12, 6))
top_cities.plot(kind='bar', color='pink')
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
#For Top 10 Cuisines by Average Cost
avg_cost_cuisines = df.groupby('Cuisines')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
print("Top 10 Cuisines by Average Cost:\n", avg_cost_cuisines)

plt.figure(figsize=(10, 6))
avg_cost_cuisines.plot(kind='bar', color='purple')
plt.title('Top 10 Cuisines by Average Cost for Two')
plt.xlabel('Cuisine')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()

```
